```python
import matplotlib.pyplot as plt

def plot_roc_curve(fpr, tpr, label=None, color=None):

    plt.plot([0, 1], [0, 1], 'k--')  # ROC
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.0])
    plt.xlabel('False Positive Rate (FPR)')
    plt.ylabel('True Positive Rate (TPR)')
    plt.title('ROC Curves')

    step_x = list(fpr)
    step_x.append(1.0)  # Add endpoint for the last step
    step_y = list(tpr)
    step_y.append(tpr[-1])  # Add endpoint for the last step
    
    plt.step(step_x, step_y, linewidth=2, label=label, color = color, alpha=0.7)
    
    plt.legend(loc="lower right")
    
fpr1 = 1-FPR_svm.to_numpy().flatten()
tpr1 = 1-FNR_svm.to_numpy().flatten()
auc_value_svm_1 = auc_value_svm.values.flatten()[0]

fpr2 = 1-FPR_rf.to_numpy().flatten()
tpr2 = 1-FNR_rf.to_numpy().flatten()
auc_value_rf_1 = auc_value_rf.values.flatten()[0]

fpr3 = 1-FPR_pcaNNet.to_numpy().flatten()
tpr3 = 1-FNR_pcaNNet.to_numpy().flatten()
auc_value_pcaNNet_1 = auc_value_pcaNNet.values.flatten()[0]

fpr4 = 1-FPR_nnet.to_numpy().flatten()
tpr4 = 1-FNR_nnet.to_numpy().flatten()
auc_value_nnet_1 = auc_value_nnet.values.flatten()[0]

fpr5 = 1-FPR_nb.to_numpy().flatten()
tpr5 = 1-FNR_nb.to_numpy().flatten()
auc_value_nb_1 = auc_value_nb.values.flatten()[0]

fpr6 = 1-FPR_knn.to_numpy().flatten()
tpr6 = 1-FNR_knn.to_numpy().flatten()
auc_value_knn_1 = auc_value_knn.values.flatten()[0]

fpr7 = 1-FPR_GBM.to_numpy().flatten()
tpr7 = 1-FNR_GBM.to_numpy().flatten()
auc_value_GBM_1 = auc_value_GBM.values.flatten()[0]

fpr8 = 1-FPR_ctree.to_numpy().flatten()
tpr8 = 1-FNR_ctree.to_numpy().flatten()
auc_value_ctree_1 = auc_value_ctree.values.flatten()[0]

plot_roc_curve(fpr1, tpr1, label=f'Support Vector Machine (AUC = {auc_value_svm_1:.4f})', color='#0072BD')

plot_roc_curve(fpr2, tpr2, label=f'Random Forest (AUC = {auc_value_rf_1:.4f})', color='#D95319')

plot_roc_curve(fpr3, tpr3, label=f'PCA Neural Network (AUC = {auc_value_pcaNNet_1:.4f})', color='#EDB120')

plot_roc_curve(fpr4, tpr4, label=f'Neural Network (AUC = {auc_value_nnet_1:.4f})', color='#7E2F8F')

plot_roc_curve(fpr5, tpr5, label=f'Naive Bayers (AUC = {auc_value_nb_1:.4f})', color='#77AC30')

plot_roc_curve(fpr6, tpr6, label=f'K Nearest Neighbor (AUC = {auc_value_knn_1:.4f})', color='#4DBEEE')

plot_roc_curve(fpr7, tpr7, label=f'Gradient Boosting Machine (AUC = {auc_value_GBM_1:.4f})', color= 'r') #'#A2142F'

plot_roc_curve(fpr8, tpr8, label=f'Conditional Inference Tree (AUC = {auc_value_ctree_1:.4f})', color= '#8EB8DC')

plt.savefig('combined_5ROC_seq.pdf', format='pdf')

plt.show()
```
